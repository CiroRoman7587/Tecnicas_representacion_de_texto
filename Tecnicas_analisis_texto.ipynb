```python
from sklearn.feature_extraction.text import CountVectorizer
import gensim
from gensim.models import Word2Vec
from gensim.models import Doc2Vec
from gensim.models.doc2vec import TaggedDocument
```

```python

documents = [


     "tres tristes tigres, trabajan perro trigal en el en el en el  trigal",
  "tragaban trigo en un trigal",
    "en tres tristes trastos",
    "tragaban trigo tres tristes tigres ",
   "El perro ladra perro perro perro perro perro perro perro perro",
    "El gato maúlla",
    "El perro y el gato yo yo yo yo yo yo son amigos",
    "El perro ladra y el gato maúlla mm m m mm mm mm mm mm"


]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)
print("Vocabulario:", vectorizer.vocabulary_)
print("Representación BoW:\n", X.toarray())
```

```python
#  EL  código genera una representación numérica de los documentos basada
#  en el conteo de palabras, lo cual es útil para muchas tareas de
# procesamiento de lenguaje natural, como clasificación de texto o #
#análisis de similitud entre documentos.

# No cuenta las letras unicas, como la (m) e (y)
# Las palabras les asigna un numero, siendo el 0 como la palabra con menos
# repeticiones en el documento.
```

```python
# Ejemplo de corpus de documentos


sentences = [
    ['el', 'perro', 'ladra'],
    ['el', 'gato', 'maúlla'],
    ['el', 'perro', 'y', 'el', 'gato', 'son', 'amigos'],
    ['el', 'perro', 'ladra', 'y', 'el', 'gato', 'maúlla']
]

# Entrenar el modelo Word2Vec
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

# Obtener el vector de la palabra 'perro'
vector = model.wv['gato']
print( "Vector de 'gato':\n", vector, "de Tamaño ", vector.shape[0])


# Encontrar palabras similares a 'perro'
similar_words = model.wv.most_similar('gato')
print("Palabras similares a 'gato':\n", similar_words)
```

```python
# La lista de palabras similares a 'gato' muestra las palabras más cercanas
# en el espacio vectorial, junto con sus puntuaciones de similitud
# (coseno de similitud). Aquí están las palabras encontradas y sus
# puntuaciones:

# 1.-'y' (0.06797593832015991): Esta es la palabra más similar a 'gato' según
# el modelo, con una similitud positiva de aproximadamente 0.068.

# 2.-'ladra' (0.009391162544488907): También tiene una similitud positiva
# pero mucho menor, cerca de 0.009.

#3.-'maúlla' (0.0045030261389911175): Tiene una similitud positiva pequeña.
# 'el' (-0.010839177295565605): Tiene una similitud negativa pequeña.
```

```python
# Ejemplo de corpus de documentos
documents = [
    TaggedDocument(words=['el', 'perro', 'ladra'], tags=['doc1']),
    TaggedDocument(words=['el', 'gato', 'maúlla'], tags=['doc2']),
    TaggedDocument(words=['el', 'perro', 'y', 'el', 'gato', 'son', 'amigos'], tags=['doc3']),
    TaggedDocument(words=['el', 'perro', 'ladra', 'y', 'el', 'gato', 'maúlla'], tags=['doc4'])
]

model = Doc2Vec(documents, vector_size=100, window=5, min_count=1, workers=4)

vector = model.dv['doc1']
print("Vector de 'doc1':\n", vector, "de Tamaño ", vector.shape[0])

similar_docs = model.dv.most_similar('doc1')
print("Documentos similares a 'doc1':\n", similar_docs)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
nltk.download('punkt')

from textblob import TextBlob
```

```python
# Definir el conjunto de documentos
documents = [
    "Data science is an interdisciplinary field",
    "Machine learning is a part of data science",
    "Data science involves statistics",
    "What is Artificial Intelligence",
    "data Artificial intelligence is a branch of computer science",
    " Big data"
]

# Crear el vectorizador TF-IDF
vectorizer = TfidfVectorizer()

# Ajustar y transformar los documentos
tfidf_matrix = vectorizer.fit_transform(documents)

# Crear un DataFrame para visualizar los resultados
df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
print(df)
```

```python
# Los valores en el DataFrame son los puntajes TF-IDF que indican la
# importancia de cada término en cada documento.

# Las nuevas palabras se colocan en una nueva linea.
# Hay que remover los acentos
```

```python
import nltk
nltk.download('punkt')
from textblob import TextBlob
```

```python

texts = [

       " It is one of the best video services",
     "Poor logistics",
    "Amazon prime video Scam",
       "Amazon prime video is a scam lt is horrible",
    "The app is horrible",
    "Terrible",
    "They solved the problem but it took too long",
    "The worst airline",
    "It is terrible, they delay flights a lot…",
    "The biggest piece of crap airline in…",
    "I want to give a thousand thanks to the whole team…",
    "I love this product! It's absolutely amazing.",
    "This is the worst experience I've ever had.",
    "The movie was okay, not great but not terrible either."
    "I'm extremely happy with the service.",
    "I hate waiting in long lines.",
    "   stinks",
   "   i love",
       " incredible  ",
       " The product that arrived is what I requested. ",
      "the product that arrived is what I requested and is in excellent condition."

]

def get_sentiment(text):
    analysis = TextBlob(text)

    return analysis.sentiment
for text in texts:
    sentiment = get_sentiment(text)
    print(f'Text: "{text}"')
    print(f'Polarity: {sentiment.polarity}, Subjectivity: {sentiment.subjectivity}')
    print()
```
